```python
import pandas as pd
```

```python
def getStock(symbol, start, end):
    df = pd.read_csv(symbol + '.csv');
    df = df[['Adj Close']]
    df.columns = df.columns + '_' + symbol
    df['Return_{}'.format(symbol)] = df['Adj Close_{}'.format(symbol)].pct_change()
     
    return df
```

```python
def getStockData(start_string, end_string):
    start = pd.to_datetime(start_string)
    end   = pd.to_datetime(end_string)
    nasdaq = getStock('^IXIC', start, end)
    frankfurt = getStock('^GDAXI', start, end)
    
    return [nasdaq, frankfurt]
```

```python
datasets = getStockData('2010-1-1', '2018-1-1');datasets
```

```python
def addFeatures(dataframe, adjclose, returns, n):
    return_n = adjclose[9:] + "Time" + str(n)
    dataframe[return_n] = dataframe[adjclose].pct_change(n)
    
    roll_n = returns[7:] + "RolMean" + str(n)
    dataframe[roll_n] = dataframe[returns].rolling(window=3)
```

```python
def applyRollingMeanDelayReturn(datasets, delta):
    for dataset in datasets:
        columns = dataset.columns
        adjclose = columns[-2]
        returns = columns[-1]
        for n in delta:
            addFeatures(dataset, adjclose, returns, n)
            
    return datasets
```

```python
datasets = applyRollingMeanDelayReturn(datasets, [3,4,5])
```
