## compute SPY daily returns & turn them into a binary variable (up, down)

```python
df1 = AXJO[(AXJO.index> '2010-01-01') & (AXJO.index < '2010-01-31')];df1
```

```python
df2 = HSI.loc[(HSI.index> '2010-01-01') & (HSI.index < '2010-01-31'),['Adj Close']]
```

```python
for i in range(0,5):
     df2["Lag_" + str(i+1)] =  df2["Adj Close"].shift(i+1)
```

```python
df2.head()
```

```python
df2['Returns'] = df2['Adj Close'].pct_change()
```

```python
df2.head(5)
```

```python
for i in range(0,5):
    df2["Lag_" + str(i+1)] =df2["Lag_" + str(i+1)].pct_change()
```

```python
df2.fillna(0, inplace=True);df2.head()
```

```python
df2.loc[(df2['Returns'] == 0), 'Returns'] = -0.000001;df2.head()
```

```python
df2["Direction"] = np.sign(df2["Returns"]);df2.head(10)
```

```python
lags = 5
lag_cols = []
for i in range(lags):
    lag_cols.append("Lag_{}".format(str(i+1)))
X = df2[lag_cols]
y = df2[["Direction"]]
```

```python
X.head()
y.head()
```

```python
start_test = '2010-01-20'
X_train = X.loc[X.index < start_test, :]
X_test = X.loc[X.index > start_test, :]
y_train = y.loc[y.index < start_test, :]
y_test  = y.loc[y.index > start_test, :]
```

```python
lr = LogisticRegression()
X_train
y_train
#y_train.values.ravel()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
y_test
```

```python
hit_rate = (sum(y_pred == y_test.iloc[:,0]))/len(y_pred)
print('Logistic Regression {:.4f}'.format(hit_rate))   
```
